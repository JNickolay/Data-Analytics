# Analysis of Classic Rock

This analysis explores various characteristics of classic rock songs, focusing on elements that define their popularity, energy, mood, and historical trends. By leveraging a dataset of classic rock tracks, the project aims to uncover key insights into what makes these songs memorable and beloved. The analysis includes identifying the top artists by popularity, highlighting the most energetic songs, examining the relationship between danceability and popularity, exploring the distribution of mood (valence) and tempo, and analyzing how the popularity of classic rock has evolved over the years. Through a combination of data analysis and visualizations, this study provides a comprehensive look into the timeless appeal of classic rock music.

## Data Analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/Users/joshuanickolay/Desktop/Development/GitHub Projects/Classic Rock Analysis/UltimateClassicRock.csv')
```

### Top Artists by Popularity
Groups the data by artist to calculate and display the most popular artists.

```python
# 1. Top Artists by Popularity
top_artists = data.groupby('Artist')['Popularity'].mean().reset_index().sort_values(by='Popularity', ascending=False).head(10)
print(top_artists)
```

### Most Energetic Songs
Identifies the top songs with the highest energy scores.

```python
# 2. Most Energetic Songs
most_energetic_songs = data[['Track', 'Artist', 'Energy']].sort_values(by='Energy', ascending=False).head(10)
print(most_energetic_songs)
```

### Danceability vs. Popularity Visualization
Uses a scatter plot to explore the relationship between danceability and popularity.

```python
# 3. Danceability vs. Popularity
plt.figure(figsize=(14, 8))
sns.scatterplot(
    x='Danceability', 
    y='Popularity', 
    data=data, 
    hue='Energy', 
    palette='viridis', 
    size='Energy', 
    sizes=(20, 200)
)
plt.title('Danceability vs. Popularity')
plt.xlabel('Danceability')
plt.ylabel('Popularity')
plt.legend(title='Energy Level')
plt.show()
```

### Valence and Tempo Distribution
Creates histograms to visualize the distribution of mood (valence) and tempo among the tracks.

```python
# 4. Valence and Tempo Distribution
plt.figure(figsize=(14, 6))
sns.histplot(data['Valence'], kde=True, color='skyblue', label='Valence', bins=20)
sns.histplot(data['Tempo'], kde=True, color='salmon', label='Tempo', bins=20)
plt.title('Distribution of Valence and Tempo')
plt.xlabel('Value')
plt.legend()
plt.show()
```

### Yearly Popularity Trends
Analyzes how the popularity of classic rock songs has changed over time using a line plot.

```python
# 5. Yearly Popularity Trends
yearly_popularity = data.groupby('Year')['Popularity'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(x='Year', y='Popularity', data=yearly_popularity, marker='o')
plt.title('Yearly Popularity Trends in Classic Rock')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(True)
plt.show()
```
