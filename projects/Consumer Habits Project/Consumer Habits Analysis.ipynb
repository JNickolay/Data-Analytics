# Consumer Habits Analysis

This analysis explores consumer purchasing behaviors by examining a dataset of consumer habits. The study focuses on several key aspects to understand what drives purchasing decisions and spending patterns. First, it identifies the most frequently purchased product types, highlighting consumer preferences. Next, it analyzes payment method preferences across different genders, revealing how these choices vary. The impact of loyalty membership on consumer spending is also assessed to determine whether members tend to spend more than non-members. Additionally, the analysis investigates factors contributing to order cancellations, identifying patterns based on product type and payment method. Finally, the study examines the correlation between age, gender, and spending, offering insights into how demographic factors influence purchasing behavior. The findings are visualized through various plots, providing a comprehensive view of the factors shaping consumer habits.

## Dataset Analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Update the path if needed
data = pd.read_csv('/Users/joshuanickolay/Desktop/Consumer Habits.csv')

# Analysis 1: Purchase Frequency by Product Type
product_preferences = data['Product Type'].value_counts().reset_index()
product_preferences.columns = ['Product Type', 'Frequency']

# Analysis 2: Payment Method Preferences by Gender
payment_by_gender = data.groupby(['Gender', 'Payment Method']).size().unstack().fillna(0)

# Analysis 3: Loyalty Membership Impact on Spending
loyalty_behavior = data.groupby('Loyalty Member')['Total Price'].mean().reset_index()

# Analysis 4: Factors Contributing to Order Cancellations
cancellation_analysis = data[data['Order Status'] == 'Cancelled'].groupby(['Product Type', 'Payment Method']).size().reset_index(name='Cancellations')

# Analysis 5: Correlation Between Age, Gender, and Spending
spending_by_demographics = data.groupby(['Age', 'Gender'])['Total Price'].mean().reset_index()
```

## Data Visualization

### Purchase Frequency by Product Type
This bar plot highlights the most frequently purchased product types, helping businesses understand consumer preferences and optimize their inventory and marketing strategies accordingly.


```python
# Visualization 1: Purchase Frequency by Product Type
sns.barplot(x='Product Type', y='Frequency', data=product_preferences, palette='viridis')
plt.title('Purchase Frequency by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Payment Method Preferences by Gender
The heatmap shows the preferred payment methods by gender, offering insights into gender-specific payment behaviors that can guide targeted payment options and promotions.

```python
# Visualization 2: Payment Method Preferences by Gender
sns.heatmap(payment_by_gender, annot=True, fmt='g', cmap='coolwarm')
plt.title('Payment Method Preferences by Gender')
plt.xlabel('Payment Method')
plt.ylabel('Gender')
plt.show()
```

### Loyalty Membership Impact on Spending 
This bar plot illustrates the average spending by loyalty program members versus non-members, showcasing the financial benefits of loyalty programs and helping to evaluate their effectiveness.

```python
# Visualization 3: Loyalty Membership Impact on Spending
sns.barplot(x='Loyalty Member', y='Total Price', data=loyalty_behavior, palette='plasma')
plt.title('Average Spending by Loyalty Membership Status')
plt.xlabel('Loyalty Member')
plt.ylabel('Average Spending')
plt.show()
```

### Cancellations by Product Type and Payment Method
This visualization identifies which product types and payment methods are most associated with cancellations, enabling businesses to address potential issues in these areas to reduce order cancellations.

```python
# Visualization 4: Factors Contributing to Order Cancellations
sns.barplot(x='Product Type', y='Cancellations', hue='Payment Method', data=cancellation_analysis, palette='magma')
plt.title('Cancellations by Product Type and Payment Method')
plt.xlabel('Product Type')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=45)
plt.show()
```

### Spending Patterns by Age and Gender
The scatter plot reveals how spending varies across different ages and genders, providing valuable demographic insights that can inform personalized marketing and pricing strategies.

```python
# Visualization 5: Correlation Between Age, Gender, and Spending
sns.scatterplot(x='Age', y='Total Price', hue='Gender', data=spending_by_demographics, palette='cool')
plt.title('Spending Patterns by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Average Spending')
plt.show()
```
